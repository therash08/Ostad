### 1️⃣ Extracting Country Code from Phone Number (Slicing)

```python
phone = "+8801712345678"
country_code=phone[0:4]
real_number=phone[4:]
print(country_code,real_number)
```

```python
phone = "+8801712345678"
country_code = phone[:4] # First slice
number = phone[4:]  # 2nd slice
print(country_code, number)
```

#### Slicing ([:]) is used to extract fixed parts of structured data.

## 4. While Loop: ATM PIN System

```python
# correct_pin = "1234"
# attempts = 0 # 3 times attempt , Maximum thereshold =3
# 0 1 2 3
# while attempts < 3 :
```

```python
correct_pin = "1234"
attempts = 0 # 3 times attempt , Maximum thereshold =3
while attempts < 3:   #0 < 3 # (True) # 2       3
    pin = input("Enter PIN: ")
    if pin == correct_pin:
        print("Access Granted")
        break #
    attempts += 1
else:
    print("Card Blocked")

print("New logic start from here")

```
